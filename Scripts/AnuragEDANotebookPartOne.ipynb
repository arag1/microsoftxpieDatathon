# <u>Looking into Factors that affect Escalation of Tickets</u>

```python
from text_analytics import *
import pandas as pd
```

```python
data = pd.read_csv('../Data/support_ticket_data.csv', encoding='Windows-1252')
data
```

```python
def text_json(text):
    
    documents = {"documents": [
        {"id": "1", "language": "en",
            "text": text}
    ]}
    return documents

def sentiment(documents):
    sentiments = sentiment_analysis_example(documents)
    x = list(map(lambda x: x["sentiment"], sentiments['documents']))
    return x

```

```python
arr = []
for x in data['Text']:
    a = text_json(x)
    arr.append(sentiment(a))
data['Sentiment'] = arr
```

```python
data
```

```python
arr = []
for x in data['Sentiment']:
    string = ""
    arr.append(string.join(x))
```

```python
data['Sentiment'] = arr
```

```python
data
```

```python
sentiment_count = data.groupby(['Sentiment']).count()[['SupportTicketID']] 
sentiment_count['SupportTicketID'] = sentiment_count['SupportTicketID']/sum(sentiment_count['SupportTicketID']) * 100
sentiment_count
```
