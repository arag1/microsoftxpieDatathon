# <u>Using Text Analysis Azure API for processing support ticket text and categorizing kind of sentiment</u>

```python
def text_json(text):
    
    documents = {"documents": [
        {"id": "1", "language": "en",
            "text": text}
    ]}
    return documents

def sentiment(documents):
    sentiments = sentiment_analysis_example(documents)
    x = list(map(lambda x: x["sentiment"], sentiments['documents']))
    return x

```

Defining methods for text analysis.

```python
arr = []
for x in data['Text']:
    a = text_json(x)
    arr.append(sentiment(a))
data['Sentiment'] = arr
```

Extracting ACTUAL sentiment types.

```python
arr = []
for x in data['Sentiment']:
    string = ""
    arr.append(string.join(x))
```

```python
data['Sentiment'] = arr
```

Converting data type of Sentiment from "array" to "string" and creating new column of dataframe as sentiment.

```python
data['DateCompleted'] = pd.to_datetime(data['DateCompleted'])
data['DateCreated'] = pd.to_datetime(data['DateCreated'])
data['Process Time'] = (data['DateCompleted'] - data['DateCreated']).dt.days
data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
normal_process = data[data['Process Time'] >= 0]
a = normal_process[normal_process['Escalated'] == 0].groupby(['Sentiment']).mean()[['Process Time']].reset_index()
b = normal_process[normal_process['Escalated'] == 1].groupby(['Sentiment']).mean()[['Process Time']].reset_index()
```

```python
sns.barplot('Sentiment', 'Process Time', data=a)
plt.title('Average Processing Time when NOT ESCALATED')
```

```python
sns.barplot('Sentiment', 'Process Time', data=b)
plt.title('Average Processing Time when ESCALATED')
```

```python
sns.boxplot(x='Sentiment', y='Process Time', data=normal_process)
```

```python
x = normal_process[normal_process['Escalated'] == 0].groupby(['Theme', 'Sentiment']).mean()[['Process Time']].reset_index()
sns.barplot('Sentiment', 'Process Time', hue="Theme", data=x)
plt.title("Average Processing Time when NOT ESCALATED")
```

```python
y = normal_process[normal_process['Escalated'] == 1].groupby(['Theme', 'Sentiment']).mean()[['Process Time']].reset_index()
sns.barplot('Sentiment', 'Process Time', hue="Theme", data=y)
plt.title("Average Processing Time when ESCALATED")
```
